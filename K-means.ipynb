# K-means Clustering



# Importing Libraries

```python
%matplotlib inline
from copy import deepcopy
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
```

# Importing the dataset for performing K-means Clustering

```python
data = pd.read_csv('xclara.csv')
print(data.shape)
data.head()
```

# Getting the values and plotting it

```python
v1 = data['V1'].values
v2 = data['V2'].values
X = np.array(list(zip(v1, v2)))
plt.scatter(v1, v2, c='black', s=7)
```

# Euclidean Distance Caculator

```python

def dist(a, b, ax=1):
    return np.linalg.norm(a - b, axis=ax)
```

# Starting with 2 random clusters

```python
# Number of clusters
k = 2
# X coordinates of random centroids
K_x = np.random.randint(0, np.max(X)-20, size=k)
# Y coordinates of random centroids
K_y = np.random.randint(0, np.max(X)-20, size=k)
K = np.array(list(zip(K_x, K_y)), dtype=np.float32)
print(K)
```

```python
# Plotting along with the Centroids
plt.scatter(v1, v2, c='#050505', s=7)
plt.scatter(K_x, K_y, marker='*', s=200, c='g')
```

```python
# To store the value of centroids when it updates
K_old = np.zeros(K.shape)
# Cluster Lables(0, 1, 2)
clusters = np.zeros(len(X))
# Error func. - Distance between new centroids and old centroids
error = dist(K, K_old, None)
# Loop will run till the error becomes zero
while error != 0:
    # Assigning each value to its closest cluster
    for i in range(len(X)):
        distances = dist(X[i], K)
        cluster = np.argmin(distances)
        clusters[i] = cluster
    # Storing the old centroid values
    K_old = deepcopy(K)
    # Finding the new centroids by taking the average value
    for i in range(k):
        points = [X[j] for j in range(len(X)) if clusters[j] == i]
        K[i] = np.mean(points, axis=0)
    error = dist(K, K_old, None)
```

# Plot with 2 clusters

```python
colors = ['r', 'g', 'b', 'y', 'c', 'm']
fig, ax = plt.subplots()
for i in range(k):
        points = np.array([X[j] for j in range(len(X)) if clusters[j] == i])
        ax.scatter(points[:, 0], points[:, 1], s=7, c=colors[i])
ax.scatter(K[:, 0], K[:, 1], marker='*', s=200, c='#050505')
```
